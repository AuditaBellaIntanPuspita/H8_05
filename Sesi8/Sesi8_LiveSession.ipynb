```python
jakarta_map = folium.Map(world_map = folium.Map(location=[-6.2, 106.8456], zoom_start=12))

#Marker
#Monas
folium.Marker([-6.175392, 106.827153],
              popup= "<i>Monas<i>",
              tooltip= "Ini Monumen Nasional"
              ).add_to(jakarta_map)
#kota tua
folium.Marker([-6.1376, 106.8171],
              popup= "Kota Tua Jakarta",
              tooltip= 'Mari Ke Kota Tua Jakarta',
              icon=folium.Icon(color='red', icon="info-sign")
              ).add_to(jakarta_map)

#Blok M
folium.CircleMarker([-6.2446, 106.8006],
                    radius=30,
                    popup= 'Blok M',
                    tooltip= 'Pandemi selesai, ayo nongkrong di sini',
                    color = '#3186cc',
                    fill = True,
                    fill_color = '#3186cc'
                    ).add_to(jakarta_map)
jakarta_map
```
