### Histogram

```python
# Matlplotlib
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head(3)
```

```python
plt.hist(tips['total_bill'])
plt.title('Total Bill Histogram' ,size=15)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
# Seaborn
sns.displot(tips['tip'])
plt.title('Tip Histogram', size=20)
plt.show()
```

```python
#Pandas
tips['tip'].hist(color='green', alpha=0.4)
plt.title('Tip Histogram', size=20)
plt.show()
```

### Lineplot

```python
#Matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

flights = sns.load_dataset("flights")
flights.head(3)
```

```python
plt.figure(figsize=(8,6))
plt.plot(flights['year'][flights['month']=='January'],
         flights['passengers'][flights['month']=='January'],
         'red')
plt.title('Number of Passengers in January Yearly', size=15)
plt.xlabel('Yearly')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
#Seaborn
sns.lineplot(data=flights, x="year", y='passengers')
plt.title('Passenger Boxplot', size = 20)
plt.show()
```

```python
#Seaborn
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
plt.title('Number of Passenger every month', size=20)
plt.show()
```

```python
#Pandas
df_january = flights(flights['month']=='January')
df_january.plot.line(x='year', y='passengers')
plt.title('Number of Passengers in January', size=20)
plt.show()
```

### Scatter Plot

```python
#Matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head(3)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(tips['total_bill'], tips ['tip'])
plt.title('Scatter Plot: Total bill vs Tip', size=15)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
#Seaborn
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title('Scatter Plot: Total bill vs Tip', size=15)
plt.show()
```

```python
#Seaborn
sns.scatterplot(data=tips, x="total_bill", y= "tip", hue='day', style='time')
plt.title('Scatter Plot: Total bill vs Tip', size=15)
plt.show()
```

```python
#Pandas
tips.plot.scatter(x='total_bill', y='tip')
plt.title('Scatter Plot: Total bill vs Tip', size=15)
plt.show()
```

### Bar Plot

```python
#Matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head(3)
```

```python
tips_by_day = tips[['tip', 'day']].groupby(['day']).mean()
tips_by_day
```

```python
x = tips['day'].unique().tolist()
y= tips_by_day['tip']

plt.figure(figsize=(8,6))
plt.bar(x,y, color = ['green', 'blue', 'red', 'yellow'])
plt.title('Barplot : Average Tips Everyday', size=15)
plt.xlabel('Days')
plt.ylabel('Tip')
plt.show()
```

```python
#Seaborn
sns.barplot(data=tips, x="day", y="tip")
plt.title('Barplot : Average Tips Everyday', size=15)
plt.show
```

```python
sns.barplot(data=tips, x="day", y="tip", hue='sex')
plt.title('Barplot : Average Tips Everyday', size=15)
plt.legend(loc=0)
plt.show
```

```python
sns.barplot(data=tips, x="tip", y="day")
plt.title('Barplot : Average Tips Everyday', size=15)
plt.legend(loc=0)
plt.show
```
