```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('melb_data.csv')
df.head()
```

```python
#No1
plt.hist(df['Price'])
plt.title('Distribusi Harga rumah', size=15)
plt.show
```

```python
#2
sns.boxplot(x= df['Distance'], y=df['Regionname'])
plt.title('Distribusi Jarak ke CBD berdasarkan Nama Region', size=15)
plt.show
```

```python
#3
sns.barplot(data=df, x="Type", y="Price")
plt.title('Perbandingan Nilai Median Harga Rumah', size=15)
plt.show
```

```python
#4

sns.scatterplot(data=df, x="Price", y="Distance", hue='Type')
plt.title('Korelasi Harga Rumah dan Jarak Rumah ke CBD', size=15)
plt.show
```

```python
# Konversi kolom "Date" agar sesuai format
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Ekstraksi data bulan dari kolom "Date" dan buat kolom "Month"
df['Month'] = df['Date'].dt.month
```

```python
import matplotlib.pyplot as plt

# Membuat Line Plot untuk tren harga rumah per bulan
plt.figure(figsize=(12, 8))

# Loop untuk setiap tipe rumah
for house_type in df['Type'].unique():
    subset = df[df['Type'] == house_type]
    monthly_median = subset.groupby('Month')['Price'].median()
    plt.plot(monthly_median.index, monthly_median.values, label=house_type)

# Menambahkan judul
plt.title('Tren Harga Rumah per Bulan berdasarkan Tipe Rumah', size=15)

# Label sumbu x dan y
plt.xlabel('Bulan')
plt.ylabel('Median Harga Rumah')

# Menambahkan legenda
plt.legend()

# Menampilkan Line Plot
plt.show()
```
